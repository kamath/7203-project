```python
# Import important packages
import numpy as np
from sklearn.datasets.samples_generator import make_circles, make_moons
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from matplotlib import pyplot as plt
import networkx as nx
import pandas as pd
```

```python
# Concentric Circles
X, clusters = make_circles(n_samples=1000, noise=.05, factor=.5, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
# Construct Adjacency Matrix
def construct_graph(X, sigma, vectorizer='scaled'):
    neighborhood = np.vectorize(lambda x: 1 if x <= sigma else 0)
    scaled_weights = np.vectorize(lambda x: np.exp(-x**2 / (2 * sigma**2)))
    W = pairwise_distances(X, metric="euclidean")
    if vectorizer == 'neighborhood':
        return np.vectorize(neighborhood)(W) #- np.eye(X.shape[0])
    else:
        return np.vectorize(scaled_weights)(W) #- np.eye(X.shape[0])

W = construct_graph(X, .1, 'neighborhood')
W, W.shape
```

```python
# Diagonalize Adjacency Matrix
def diagonalize(W):
    return np.diag(np.sum(W, axis=1))

D = diagonalize(W)
D
```

```python
# Compute Graph Laplacian with option to normalize
def laplacian(D, W, normalize=False):
    if normalize:
        return nx.normalized_laplacian_matrix(W).toarray()
    else:
        return D - W

L = laplacian(D, W)
```

```python
# Eigendecomposition of Laplacian Graph
def eigendecomp(L, keep_n = None):
    e, v = np.linalg.eig(L)
    # Sort eigenvalues from lowest to highest
    e = np.array(list(map(lambda x: x[0], sorted(enumerate(e), key=lambda x: x[1]))))
    v = np.vstack([v[:,i] for i in e[:keep_n]]).T
    print('eigenvalues:')
    print(e.shape)
    # eigenvectors
    print('eigenvectors:')
    print(v.shape)
    return v
    
print(W.shape)
phi = eigendecomp(L, 3)
```

```python
# Plot results
plt.scatter(X[:,0], X[:,1], c=KMeans(2).fit(phi).predict(phi))
```

```python
# Start to finish
def end_to_end(X, sigma=.1, vectorizer='scaled', normalize_laplacian=False, eigenvalue_n=None, clusters=2):
    W = construct_graph(X, sigma, vectorizer)
    D = diagonalize(W)
    L = laplacian(D, W, normalize_laplacian)
    phi = eigendecomp(L, eigenvalue_n)
    return KMeans(clusters).fit(phi).predict(phi)
```

```python
# Plot difference vs KMeans
def plot_vs_kmeans(X, n_clusters=2):
    plt.scatter(X[:,0], X[:,1], c=KMeans(2).fit(X).predict(X))
    plt.title("KMeans")
    plt.show()
    plt.scatter(X[:,0], X[:,1], c=end_to_end(X, eigenvalue_n=2))
    plt.title("Spectral Clustering")
    plt.show()
```

```python
X, clusters = make_circles(n_samples=1000, noise=.05, factor=.5, random_state=0)
plot_vs_kmeans(X)
```

```python
# Write data to CSV for website
pd.DataFrame({"x": X[:, 0], "y": X[:, 1], "rightColor": end_to_end(X, eigenvalue_n=2),
              "badColor": KMeans(2).fit(X).predict(X)}).to_csv("concentric_circles.csv")
```

```python
# NBA Analysis - get matchup data for all players
from nba_api.stats.endpoints import leagueseasonmatchups

df = leagueseasonmatchups.LeagueSeasonMatchups(season='2019-20').get_data_frames()[0]
print(len(set(list(df['OFF_PLAYER_ID'].unique()) + list(df['DEF_PLAYER_ID'].unique()))))
print(df.columns)
df
```

```python
df.to_csv("players.csv")
```

```python
# Get time in seconds on average per game
df['TIME'] = df['MATCHUP_MIN'].apply(lambda x: int(x.split(':')[0]) * 60 + float(x.split(':')[1])) / df['GP']

# Keep only relevant columns
adj = df[['OFF_PLAYER_ID', 'DEF_PLAYER_ID', 'TIME']]

# Make the graph symmetric by combining offensive/defensive matchups
bottom = df[['OFF_PLAYER_ID', 'DEF_PLAYER_ID', 'TIME']].rename(columns={"OFF_PLAYER_ID": "DEF_PLAYER_ID", 
                                                                             "DEF_PLAYER_ID": "OFF_PLAYER_ID"})
adj = pd.concat([adj, bottom]).groupby(['OFF_PLAYER_ID', 'DEF_PLAYER_ID']).sum().reset_index()

# Keep only matchups longer than 600 seconds
adj = adj.loc[adj['TIME'] > 600]
# Normalize Time
adj['weight'] = adj['TIME'] / np.linalg.norm(adj['TIME'])

player_graph = nx.from_pandas_edgelist(
    adj,
    'DEF_PLAYER_ID',
    'OFF_PLAYER_ID',
    'weight'
)
W = np.array(nx.to_numpy_matrix(player_graph))
print(W.shape)
adj
```

```python
# Plot results
D = diagonalize(W)
L = laplacian(D, W, False)
phi = eigendecomp(L, 5)
k = 5
preds = KMeans(k).fit(phi).predict(phi)
plt.scatter(phi[:, -2], phi[:, -1], c=preds)
np.unique(preds, return_counts=True)
```

```python
# Persist data for website

from nba_api.stats.static import players
nba_players = players.get_players()
nba_players = {a['id']: a for a in nba_players}
player_data = [nba_players[a]['full_name'] for a in player_graph.nodes]

# Compute centrality for easy sorting to validate data
centrality = nx.degree_centrality(player_graph)

player_table = pd.DataFrame({"player": player_data, 'cluster': preds, "X": phi[:, -2], "Y": phi[:, -1], 'centrality': [centrality[a] for a in player_graph.nodes]})
player_table = player_table.sort_values('centrality')[::-1]
for n in np.unique(preds):
    print(n)
    display(player_table[player_table['cluster'] == n].head())
```
